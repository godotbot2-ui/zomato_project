## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Check dataset shape
print(data1.shape, data2.shape)
````

````python
# Check data types and missing values
print(data1.info(), data2.info())
````

````python
# Remove duplicate rows
data1 = data1.drop_duplicates()
data2 = data2.drop_duplicates()
````

````python
# Check missing values
print(data1.isnull().sum(), data2.isnull().sum())
````

````python
# Handle missing values
# (Drop columns with too many missing values if any)
threshold1 = 0.5 * len(data1)
data1 = data1.dropna(axis=1, thresh=threshold1)

threshold2 = 0.5 * len(data2)
data2 = data2.dropna(axis=1, thresh=threshold2)
````

````python
for col in data1.select_dtypes(include='object').columns:
    data1[col].fillna(data1[col].mode()[0], inplace=True)

for col in data1.select_dtypes(include='number').columns:
    data1[col].fillna(data1[col].median(), inplace=True)

for col in data2.select_dtypes(include='object').columns:
    data2[col].fillna(data2[col].mode()[0], inplace=True)

for col in data2.select_dtypes(include='number').columns:
    data2[col].fillna(data2[col].median(), inplace=True)
````

````python
import pandas as pd

meta = pd.read_csv("Zomato Restaurant names and Metadata.csv")
reviews = pd.read_csv("Zomato Restaurant reviews.csv")

# Merge on restaurant name
merged_data = pd.merge(
    meta,
    reviews,
    left_on="Name",
    right_on="Restaurant",
    how="inner"
)
````

````python
# Remove currency symbol and text
merged_data['Cost'] = (
    merged_data['Cost']
    .str.replace('₹', '', regex=False)
    .str.replace('for two', '', regex=False)
    .str.replace(',', '', regex=False)
    .str.strip()
)

merged_data['Cost'] = pd.to_numeric(merged_data['Cost'], errors='coerce')
````

````python
merged_data['Rating'] = pd.to_numeric(merged_data['Rating'], errors='coerce')
merged_data = merged_data.dropna(subset=['Cost', 'Rating'])
````

````python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.scatter(merged_data['Cost'], merged_data['Rating'], alpha=0.5)
plt.xlabel("Cost for Two (INR)")
plt.ylabel("Restaurant Rating")
plt.title("Cost vs Restaurant Ratings")
plt.show()
````

##### 1. Why did you pick the specific chart?

A scatter plot was chosen to examine the relationship between restaurant pricing and customer ratings. Since both cost and rating are numerical variables, a scatter plot effectively reveals trends, correlations, and outliers between them.

##### 2. What is/are the insight(s) found from the chart?

The visualization indicates that higher-priced restaurants do not always receive higher ratings. Several low to moderately priced restaurants achieve good ratings, suggesting that customer satisfaction depends more on quality and service than cost alone.

##### 3. Will the gained insights help creating a positive business impact?
Yes, this insight helps restaurant owners focus on delivering better value rather than increasing prices. Affordable restaurants with high ratings can be promoted, while expensive restaurants can reassess pricing and service strategies to improve customer satisfaction.

##### Are there any insights that lead to negative growth?

##### 1. Why did you pick the specific chart?

The KDE plot shows that restaurant costs are highly concentrated in the lower to mid-price range, indicating that most restaurants target budget-conscious customers. The distribution is right-skewed, with fewer restaurants operating in the premium price segment.

### Hypothetical Statement - 2

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

Null Hypothesis (H₀):

There is no significant difference in restaurant ratings between short reviews and long reviews.

Alternate Hypothesis (H₁):

There is a significant difference in restaurant ratings between short reviews and long reviews.

#### 2. Perform an appropriate statistical test.

````python
from scipy.stats import ttest_ind
import pandas as pd

# Convert to numeric
merged_data['Rating'] = pd.to_numeric(merged_data['Rating'], errors='coerce')
merged_data['Review_Length'] = pd.to_numeric(merged_data['Review_Length'], errors='coerce')

# Drop missing values
clean_data = merged_data.dropna(subset=['Rating', 'Review_Length'])

# Create groups based on median review length
median_length = clean_data['Review_Length'].median()

short_reviews = clean_data[clean_data['Review_Length'] <= median_length]['Rating']
long_reviews = clean_data[clean_data['Review_Length'] > median_length]['Rating']

# Perform Welch's t-test
t_stat, p_value = ttest_ind(short_reviews, long_reviews, equal_var=False)

t_stat, p_value
````

##### Which statistical test have you done to obtain P-Value?

Independent Samples t-test (Welch’s t-test)

##### Why did you choose the specific statistical test?

The independent t-test was chosen because the goal was to compare the mean ratings of two independent groups (restaurants with and without online delivery), where the dependent variable is continuous.

### Hypothetical Statement - 3

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

Null Hypothesis (H₀):

There is no significant difference in restaurant ratings across different cuisine categories.

Alternate Hypothesis (H₁):

There is a significant difference in restaurant ratings across different cuisine categories.

#### 2. Perform an appropriate statistical test.

````python
from scipy.stats import f_oneway
import pandas as pd

# Convert Rating to numeric
merged_data['Rating'] = pd.to_numeric(merged_data['Rating'], errors='coerce')

# Handle cuisines (take first cuisine if multiple)
merged_data['Primary_Cuisine'] = merged_data['Cuisines'].astype(str).str.split(',').str[0]

# Drop missing values
clean_data = merged_data.dropna(subset=['Rating', 'Primary_Cuisine'])

# Keep only top cuisines to avoid noise
top_cuisines = clean_data['Primary_Cuisine'].value_counts().head(5).index
filtered_data = clean_data[clean_data['Primary_Cuisine'].isin(top_cuisines)]

# Group ratings by cuisine
groups = [
    filtered_data[filtered_data['Primary_Cuisine'] == cuisine]['Rating']
    for cuisine in top_cuisines
]

# Perform One-Way ANOVA
f_stat, p_value = f_oneway(*groups)

f_stat, p_value
````

##### Which statistical test have you done to obtain P-Value?

One-Way Analysis of Variance (ANOVA)

##### Why did you choose the specific statistical test?

One-way ANOVA was chosen because it compares the mean ratings across more than two independent cuisine categories.

### 2. Handling Outliers

````python
# Function to cap outliers using IQR
def cap_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    df[column] = df[column].clip(lower, upper)
    return df
````

````python
# Apply outlier capping
merged_data = cap_outliers(merged_data, 'Cost')
merged_data = cap_outliers(merged_data, 'Review_Length')
````

##### What all outlier treatment techniques have you used and why did you use those techniques?

The Interquartile Range (IQR) method was used to detect outliers in numerical variables such as Cost and Review_Length.

Instead of removing outliers, extreme values were capped within acceptable bounds (winsorization) to preserve data points.

This method was chosen because it is robust to skewed distributions and prevents extreme values from disproportionately influencing analysis results.

#### 1. Expand Contraction

````python
contractions = {
    "can't": "cannot", "won't": "will not", "n't": " not",
    "'re": " are", "'s": " is", "'d": " would",
    "'ll": " will", "'t": " not", "'ve": " have", "'m": " am"
}

def expand_contractions(text):
    for key, value in contractions.items():
        text = re.sub(key, value, text)
    return text

merged_data['Review'] = merged_data['Review'].astype(str).apply(expand_contractions)
````

#### 4. Removing URLs & Removing words and digits contain digits.

````python
def remove_urls_digits(text):
    text = re.sub(r'http\S+|www\S+', '', text)
    text = re.sub(r'\w*\d\w*', '', text)
    return text

merged_data['Review'] = merged_data['Review'].apply(remove_urls_digits)
````

````python
stop_words = set(stopwords.words('english'))

def remove_stopwords(text):
    words = text.split()
    words = [w for w in words if w not in stop_words]
    return " ".join(words)

merged_data['Review'] = merged_data['Review'].apply(remove_stopwords)
merged_data['Review'] = merged_data['Review'].str.strip()
````
